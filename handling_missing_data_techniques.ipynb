# HANDLING MISSING DATA

## Why missing data?

In real world data, there are some instances where a particular element is absent because of various reasons, such as, corrupt data, failure to load the information, or incomplete extraction. 

## 1. Deleting Rows

This method is advised only when there are enough samples in the data set.

Cons:<br>
Loss of information and data.<br>
Works poorly if the percentage of missing values is high (say 30%), compared to the whole dataset.<br>

## 2. Replacing With Mean/Median/Mode

This strategy can be applied on a feature which has numeric data like the age of a person or the ticket fare. We can calculate the mean, median or mode of the feature and replace it with the missing values. 

Pros:<br>
This is a better approach when the data size is small.<br>
It can prevent data loss which results in removal of the rows and columns.<br>

Cons:<br>
Imputing the approximations add variance and bias.<br>

## 3. Assigning An Unique Category

Pros:<br>
Resulting in low variance.<br>
Cons:<br>
Adds less variance<br>

## 4. Predicting The Missing Values.

Using the features which do not have missing values, we can predict the nulls with the help of a machine learning algorithm. This method may result in better accuracy, unless a missing value is expected to have a very high variance. We will be using linear regression to replace the nulls in the feature ‘age’, using other available features. One can experiment with different algorithms and check which gives the best accuracy instead of sticking to a single algorithm.



Pros:<br>
Yields unbiased estimates of the model parameters.<br>
